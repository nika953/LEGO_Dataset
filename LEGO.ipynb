```python
def fill_nan_pieces(lego: pd.DataFrame, grouping_value: str, filling_value: str) -> pd.DataFrame:
    lego = lego.copy(deep=True)
    grouped = lego.groupby(grouping_value)[filling_value].mean()
    grouped.fillna(lego[filling_value].mean(), inplace=True)

    for value in lego[grouping_value].unique():
        indexes = list(lego[(lego[grouping_value] == value) & (lego[filling_value].isna() == True)].index)
        for index in indexes:
            lego.loc[[index], filling_value] = grouped[value]
    
    return lego
```

```python
new_lego = fill_nan_pieces(lego, grouping_value='Packaging', filling_value='Pieces')
```
